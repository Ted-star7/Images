#Script that automatically generates the 52 AI image prompts for your taxonomy


```python
subcategories = sorted(final_df['POS_Subcategory'].unique())

print("Total Subcategories:", len(subcategories))
subcategories
```

```python
prompts = {}

for sub in subcategories:

    prompt = f"""
Studio product photography of {sub.lower()}, centered product representation,
clean white background, soft shadow, high-end retail catalog style,
minimalistic composition, ultra realistic, 4k, professional lighting
"""

    prompts[sub] = prompt.strip()

prompts
```

```python
prompt_df = pd.DataFrame({
    "POS_Subcategory": list(prompts.keys()),
    "AI_Image_Prompt": list(prompts.values())
})

prompt_df.head()
```

```python
prompt_df.to_csv(
"/content/drive/MyDrive/Dataset/Uzapoint_Taxonomy_Image_Prompts.csv",
index=False
)
```

```python
image_map = {
    sub: f"https://cdn.uzapoint.com/taxonomy/{sub.lower().replace(' ','_')}.jpg"
    for sub in subcategories
}

missing_mask = final_df['Product Image'].isna()

final_df.loc[missing_mask, 'Product Image'] = final_df.loc[
    missing_mask, 'POS_Subcategory'
].map(image_map)
```

```python
# ============================================
# UZAPOINT GLOBAL PRODUCT IMAGE GENERATOR
# ============================================

import os
import pandas as pd
import requests
from google.colab import drive

# Mount Drive
drive.mount('/content/drive')

# Folder to store images
image_folder = "/content/drive/MyDrive/taxonomy_images"
os.makedirs(image_folder, exist_ok=True)

# --------------------------------------------
# GET SUBCATEGORIES FROM YOUR DATA
# --------------------------------------------

subcategories = sorted(combined_df["POS_Subcategory"].unique())

print("Total Subcategories:", len(subcategories))
print(subcategories)

# --------------------------------------------
# GENERATE PLACEHOLDER IMAGES USING POLLINATION AI
# --------------------------------------------

image_paths = {}

for sub in subcategories:

    filename = sub.replace(" ", "_").replace("/", "_") + ".jpg"
    filepath = os.path.join(image_folder, filename)

    prompt = f"studio product photo of {sub}, clean white background, professional lighting, ecommerce product photography"

    url = f"https://image.pollinations.ai/prompt/{prompt}"

    try:
        img = requests.get(url).content
        with open(filepath, "wb") as f:
            f.write(img)

        image_paths[sub] = filepath
        print("Saved:", filename)

    except:
        print("Failed:", sub)

print("Image generation complete")
```

```python
github_base = "https://raw.githubusercontent.com/Ted-star7/product-images/main/images/"

image_map = {}

for sub in subcategories:

    filename = sub.replace(" ", "_").replace("/", "_") + ".jpg"
    image_map[sub] = github_base + filename

print(image_map)
```

```python
combined_df["Product Image"] = combined_df["POS_Subcategory"].map(image_map)

combined_df.head()
```
