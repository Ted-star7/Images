```python
import pandas as pd
import numpy as np
import json
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Load dataset
file_path = '/content/drive/MyDrive/Dataset/Uzapoint_Master_Clean_Data.csv'
df = pd.read_csv(file_path)

# Preview dataset
df.head()
```

```python
image_df = df[['Product ID', 'Product Image']].copy()

image_df.head()
```

```python
missing_mask = image_df['Product Image'].isna() | (image_df['Product Image'] == "")

missing_count = missing_mask.sum()
total_products = len(image_df)

print("Total Products:", total_products)
print("Missing Images:", missing_count)
print("Existing Images:", total_products - missing_count)
```

```python
import pandas as pd
import os
from google.colab import drive

drive.mount('/content/drive')

folder_path = "/content/drive/MyDrive/Dataset/category_review_simple_20260303_195522/category_review_simple_20260303_195522"

import os
files = os.listdir(folder_path)

files
```

```python
import pandas as pd
import os

dfs = []

for file in files:

    path = os.path.join(folder_path, file)

    df_temp = pd.read_csv(path)

    dfs.append(df_temp)

combined_df = pd.concat(dfs, ignore_index=True)

combined_df.head()
```

```python
print("Total Products:", len(combined_df))
```
