```python
# ============================================
# UZAPOINT GLOBAL PRODUCT IMAGE GENERATOR (FIXED)
# ============================================

import os
import pandas as pd
import requests
import urllib.parse
from google.colab import drive
import time
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

# Mount Drive (if not already mounted)
drive.mount('/content/drive')

# ============================================
# STEP 1: LOAD YOUR EXISTING DATA
# ============================================

# Load the final combined dataframe
file_path = '/content/drive/MyDrive/Dataset/Uzapoint_POS_Taxonomy_Base.csv'
combined_df = pd.read_csv(file_path)

# Load the original master data with product images
master_file_path = '/content/drive/MyDrive/Dataset/Uzapoint_Master_Clean_Data.csv'
master_df = pd.read_csv(master_file_path)
image_df = master_df[['Product ID', 'Product Image']].copy()

print("Data loaded successfully!")
print(f"Total products: {len(combined_df)}")
print(f"Total unique subcategories: {combined_df['POS_Subcategory'].nunique()}")
print(f"Total unique categories: {combined_df['Category'].nunique()}")

# ============================================
# STEP 2: CHOOSE YOUR APPROACH
# ============================================

print("\n" + "="*50)
print("CHOOSE YOUR IMAGE GENERATION APPROACH:")
print("="*50)
print("1. Generate images for SUBCATEGORIES (52 images)")
print("2. Generate images for CATEGORIES (6 images)")
print("3. Generate images for BOTH")
print("="*50)

# Set your choice here (change this number)
CHOICE = 1  # Change to 1, 2, or 3

# ============================================
# STEP 3: DEFINE IMAGE GENERATION FUNCTION WITH RETRY LOGIC
# ============================================

def generate_images_with_retry(items, item_type, folder_name, max_retries=3):
    """
    Generate images for a list of items with retry logic for timeouts
    """

    # Create folder in Google Drive
    image_folder = f"/content/drive/MyDrive/{folder_name}"
    os.makedirs(image_folder, exist_ok=True)

    # Setup session with retries
    session = requests.Session()
    retry_strategy = Retry(
        total=2,
        backoff_factor=1,
        status_forcelist=[429, 500, 502, 503, 504],
    )
    adapter = HTTPAdapter(max_retries=retry_strategy)
    session.mount("http://", adapter)
    session.mount("https://", adapter)

    print(f"\nGenerating {len(items)} {item_type} images...")
    print("-" * 40)

    # Store image paths/URLs
    image_map = {}
    failed_items = []

    for i, item in enumerate(items, 1):
        # Create filename (sanitize for filesystem)
        filename = item.replace(" ", "_").replace("/", "_").replace("&", "and")
        filename = filename.replace("(", "").replace(")", "").replace("'", "")
        filename = filename + ".jpg"
        filepath = os.path.join(image_folder, filename)

        # Create a descriptive prompt
        if item_type == 'category':
            prompt = f"professional studio photography collage of various {item} products, arranged neatly, clean white background, high-end retail catalog style, 4k quality, multiple products showcase"
        else:
            prompt = f"studio product photography of {item}, centered product representation, clean white background, soft shadow, high-end retail catalog style, minimalistic composition, ultra realistic, 4k, professional lighting"

        # URL encode the prompt
        encoded_prompt = urllib.parse.quote(prompt)

        # Pollinations AI URL
        url = f"https://image.pollinations.ai/prompt/{encoded_prompt}"

        print(f"[{i}/{len(items)}] Generating: {item}")

        # Try to download with retries
        success = False
        for attempt in range(max_retries):
            try:
                # Shorter timeout for first attempt, longer for retries
                timeout = 20 if attempt == 0 else 40
                response = session.get(url, timeout=timeout)

                if response.status_code == 200:
                    with open(filepath, 'wb') as f:
                        f.write(response.content)

                    image_map[item] = {
                        'local_path': filepath,
                        'url': url,
                        'filename': filename,
                        'prompt': prompt
                    }
                    print(f"  ✓ Saved: {filename}")
                    success = True
                    break
                else:
                    print(f"  ⚠ Attempt {attempt + 1}: HTTP {response.status_code}")

            except requests.exceptions.Timeout:
                print(f"  ⚠ Attempt {attempt + 1}: Timeout")
            except requests.exceptions.ConnectionError:
                print(f"  ⚠ Attempt {attempt + 1}: Connection error")
            except Exception as e:
                print(f"  ⚠ Attempt {attempt + 1}: {str(e)[:50]}")

            time.sleep(2)  # Wait before retry

        if not success:
            failed_items.append(item)
            print(f"  ✗ Failed after {max_retries} attempts")

        # Longer delay between different items
        time.sleep(3)

    print(f"\n✓ Completed! {len(image_map)}/{len(items)} images generated")
    if failed_items:
        print(f"✗ Failed items: {', '.join(failed_items[:5])}{'...' if len(failed_items) > 5 else ''}")
    print(f"Images saved to: {image_folder}")

    return image_map, image_folder, failed_items

# ============================================
# STEP 4: GENERATE IMAGES BASED ON CHOICE
# ============================================

all_image_maps = {}
failed_lists = {}

if CHOICE in [1, 3]:
    # Generate subcategory images
    subcategories = sorted(combined_df['POS_Subcategory'].unique())
    sub_map, sub_folder, failed_subs = generate_images_with_retry(
        subcategories, 'subcategory', 'taxonomy_subcategory_images'
    )
    all_image_maps['subcategories'] = sub_map
    failed_lists['subcategories'] = failed_subs

if CHOICE in [2, 3]:
    # Generate category images
    categories = sorted(combined_df['Category'].unique())
    cat_map, cat_folder, failed_cats = generate_images_with_retry(
        categories, 'category', 'taxonomy_category_images'
    )
    all_image_maps['categories'] = cat_map
    failed_lists['categories'] = failed_cats

# ============================================
# STEP 5: CREATE URL MAPPINGS
# ============================================

print("\n" + "="*50)
print("CREATING URL MAPPINGS")
print("="*50)

# Create mapping dictionaries
subcategory_urls = {}
category_urls = {}

if CHOICE in [1, 3]:
    for sub in subcategories:
        if sub in sub_map:
            # Use local file path
            subcategory_urls[sub] = sub_map[sub]['local_path']
        else:
            # For failed ones, create a placeholder or use a default
            subcategory_urls[sub] = "IMAGE_GENERATION_FAILED"

    # Add to dataframe
    combined_df['Generated_Subcategory_Image'] = combined_df['POS_Subcategory'].map(subcategory_urls)
    print(f"✓ Added subcategory image paths to dataframe")

if CHOICE in [2, 3]:
    for cat in categories:
        if cat in cat_map:
            category_urls[cat] = cat_map[cat]['local_path']
        else:
            category_urls[cat] = "IMAGE_GENERATION_FAILED"

    combined_df['Generated_Category_Image'] = combined_df['Category'].map(category_urls)
    print(f"✓ Added category image paths to dataframe")

# ============================================
# STEP 6: HANDLE MISSING PRODUCT IMAGES (FIXED)
# ============================================

print("\n" + "="*50)
print("HANDLING MISSING PRODUCT IMAGES")
print("="*50)

# Merge with original image data
final_df = combined_df.merge(
    image_df,
    on="Product ID",
    how="left"
)

# Check actual column names after merge
print("Columns in final_df:", final_df.columns.tolist())

# Handle missing images based on which images we generated
if CHOICE in [1, 3]:
    # Use subcategory images for missing product images
    missing_mask = final_df['Product Image'].isna()
    final_df.loc[missing_mask, 'Final_Product_Image'] = final_df.loc[missing_mask, 'Generated_Subcategory_Image']
    final_df.loc[~missing_mask, 'Final_Product_Image'] = final_df.loc[~missing_mask, 'Product Image']

    print(f"\nUsing subcategory images to fill missing product images:")
else:
    # Use category images for missing product images
    missing_mask = final_df['Product Image'].isna()
    final_df.loc[missing_mask, 'Final_Product_Image'] = final_df.loc[missing_mask, 'Generated_Category_Image']
    final_df.loc[~missing_mask, 'Final_Product_Image'] = final_df.loc[~missing_mask, 'Product Image']

    print(f"\nUsing category images to fill missing product images:")

# Statistics
total_products = len(final_df)
filled_images = missing_mask.sum()
existing_images = total_products - filled_images

print(f"Total Products: {total_products}")
print(f"Products with original images: {existing_images}")
print(f"Products needing images: {filled_images}")
print(f"Successfully filled: {filled_images - (final_df['Final_Product_Image'] == 'IMAGE_GENERATION_FAILED').sum()}")
print(f"Failed to fill: {(final_df['Final_Product_Image'] == 'IMAGE_GENERATION_FAILED').sum()}")

# ============================================
# STEP 7: SAVE FINAL DATASET
# ============================================

# Keep relevant columns
columns_to_keep = ['Product ID', 'Category', 'POS_Subcategory', 'Product Label',
                   'Product Image', 'Final_Product_Image']
if CHOICE in [1, 3]:
    columns_to_keep.append('Generated_Subcategory_Image')
if CHOICE in [2, 3]:
    columns_to_keep.append('Generated_Category_Image')

final_output = final_df[[col for col in columns_to_keep if col in final_df.columns]]

# Save the final dataset
output_path = "/content/drive/MyDrive/Dataset/Uzapoint_Final_With_Generated_Images.csv"
final_output.to_csv(output_path, index=False)

print(f"\n✓ Final dataset saved to: {output_path}")

# ============================================
# STEP 8: CREATE REFERENCE CSV FOR SUCCESSFUL IMAGES
# ============================================

# Create a reference dataframe for successfully generated images
reference_data = []

if CHOICE in [1, 3]:
    for sub, data in sub_map.items():
        reference_data.append({
            'Type': 'Subcategory',
            'Name': sub,
            'Local_Path': data['local_path'],
            'Filename': data['filename'],
            'Status': 'Success'
        })
    # Add failed ones
    for sub in failed_lists.get('subcategories', []):
        reference_data.append({
            'Type': 'Subcategory',
            'Name': sub,
            'Local_Path': 'FAILED',
            'Filename': sub.replace(" ", "_").replace("/", "_") + ".jpg",
            'Status': 'Failed'
        })

if CHOICE in [2, 3]:
    for cat, data in cat_map.items():
        reference_data.append({
            'Type': 'Category',
            'Name': cat,
            'Local_Path': data['local_path'],
            'Filename': data['filename'],
            'Status': 'Success'
        })
    # Add failed ones
    for cat in failed_lists.get('categories', []):
        reference_data.append({
            'Type': 'Category',
            'Name': cat,
            'Local_Path': 'FAILED',
            'Filename': cat.replace(" ", "_").replace("/", "_") + ".jpg",
            'Status': 'Failed'
        })

if reference_data:
    reference_df = pd.DataFrame(reference_data)
    ref_output_path = "/content/drive/MyDrive/Dataset/Generated_Images_Reference.csv"
    reference_df.to_csv(ref_output_path, index=False)
    print(f"✓ Image reference saved to: {ref_output_path}")

# ============================================
# STEP 9: SUMMARY
# ============================================

print("\n" + "="*50)
print("SUMMARY")
print("="*50)
print(f"Successfully generated images: {len(sub_map) if CHOICE in [1,3] else 0}/{len(subcategories) if CHOICE in [1,3] else 0} subcategories")
if CHOICE in [2,3]:
    print(f"Successfully generated images: {len(cat_map) if CHOICE in [2,3] else 0}/{len(categories)} categories")
print(f"\nFailed items (will need to be retried manually):")
for type_name, failed_list in failed_lists.items():
    if failed_list:
        print(f"{type_name}: {', '.join(failed_list)}")

# ============================================
# OPTIONAL: RETRY FAILED ITEMS
# ============================================

retry_choice = input("\nDo you want to retry failed items now? (yes/no): ")
if retry_choice.lower() == 'yes':
    for type_name, failed_list in failed_lists.items():
        if failed_list:
            print(f"\nRetrying {type_name}...")
            if type_name == 'subcategories':
                new_map, _, new_failed = generate_images_with_retry(
                    failed_list, 'subcategory', 'taxonomy_subcategory_images_retry'
                )
                # Update maps
                sub_map.update(new_map)
                failed_lists[type_name] = new_failed
            elif type_name == 'categories':
                new_map, _, new_failed = generate_images_with_retry(
                    failed_list, 'category', 'taxonomy_category_images_retry'
                )
                cat_map.update(new_map)
                failed_lists[type_name] = new_failed

    print("\nRetry complete! Check the reference CSV for updated status.")
```
